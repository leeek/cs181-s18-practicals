```python
import pandas as pd
```

```python
##########################
## SET UP TRAINING DATA ##
##########################

# read in featurized training data: contains eigenvalues for adjacency and coulomb matrices
df_featurized_train = pd.read_csv("data/coulomb_and_eigen_full_train.csv")

# drop row number
df_featurized_train = df_featurized_train.drop(['Unnamed: 0'], axis=1)

# rename smile column to smiles (for merging with csv of default features)
df_featurized_train.rename(columns={'smile' : 'smiles'}, inplace=True)

# read in original data with default features
df_train = pd.read_csv("train.csv")
df_train = pd.merge(df_featurized_train, df_train, on='smiles')

# clean up merged data
df_train = df_train.drop(['gap_x'], axis = 1)
#df_train = df_train.drop(['smiles'], axis = 1)
df_train.rename(columns={'gap_y' : 'gap'}, inplace = True)
```

```python
cols = list(df_train)
```

```python
cols.insert(-1, cols.pop(cols.index('smiles')))
```

```python
df_train = df_train.ix[:,cols]
```

```python
df_train.to_csv('merged_train_smiles.csv',index=False)
```
